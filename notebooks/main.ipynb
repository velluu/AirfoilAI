```python
# Load the saved dataset (fast!)
import joblib
import pandas as pd
import numpy as np

# Load dataset
dataset = joblib.load('../data/processed/dataset_full_train.pkl')
print(f"Dataset loaded: {len(dataset)} samples")

# Display first 20 records
print("\n" + "="*60)
print("FIRST 20 RECORDS:")
print("="*60)

for i in range(min(20, len(dataset))):
    sample = dataset[i]
    print(f"\nSample {i}:")
    
    if isinstance(sample, (list, tuple)):
        print(f"  Type: {type(sample).__name__} with {len(sample)} elements")
        for j, item in enumerate(sample):
            if hasattr(item, 'shape'):
                print(f"    Element {j}: {type(item).__name__}, shape={item.shape}")
            else:
                print(f"    Element {j}: {type(item).__name__}, value={item}")
    elif isinstance(sample, dict):
        print(f"  Keys: {list(sample.keys())}")
        for key, value in sample.items():
            if hasattr(value, 'shape'):
                print(f"    {key}: shape={value.shape}")
            else:
                print(f"    {key}: {value}")
    else:
        print(f"  {sample}")
```
